```python
train_df = pd.read_csv(
    "/content/liar_dataset/train.tsv",
    sep="\t",
    header=None,
    names=cols
)

valid_df = pd.read_csv(
    "/content/liar_dataset/valid.tsv",
    sep="\t",
    header=None,
    names=cols
)

test_df = pd.read_csv(
    "/content/liar_dataset/test.tsv",
    sep="\t",
    header=None,
    names=cols
)
```

```python
print(train_df.head())

print(train_df.shape)

print(train_df.info())
```

```python
print(train_df["label"].value_counts())
```

```python
# Labels considered as Fake
fake_labels = {"false", "pants-fire", "barely-true"}

# Function to convert to binary
def to_binary(label):
    if label in fake_labels:
        return 1    # Fake
    else:
        return 0    # Real

# Apply conversion
train_df["binary_label"] = train_df["label"].apply(to_binary)
valid_df["binary_label"] = valid_df["label"].apply(to_binary)
test_df["binary_label"] = test_df["label"].apply(to_binary)

# Check distribution
print(train_df["binary_label"].value_counts())
```

```python
train_df = train_df[["statement", "binary_label"]].rename(
    columns={
        "statement": "text",
        "binary_label": "label"
    }
)

valid_df = valid_df[["statement", "binary_label"]].rename(
    columns={
        "statement": "text",
        "binary_label": "label"
    }
)

test_df = test_df[["statement", "binary_label"]].rename(
    columns={
        "statement": "text",
        "binary_label": "label"
    }
)

train_df.head()
```

```python
print(train_df.isnull().sum())
print(valid_df.isnull().sum())
print(test_df.isnull().sum())
```

```python
train_df = train_df.dropna()
valid_df = valid_df.dropna()
test_df = test_df.dropna()
```

```python
train_df.to_csv("train_clean.csv", index=False)
valid_df.to_csv("valid_clean.csv", index=False)
test_df.to_csv("test_clean.csv", index=False)

print("✅ Preprocessed datasets saved successfully!")
```
